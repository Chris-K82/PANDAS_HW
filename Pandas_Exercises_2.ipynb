### Step 4. How many products cost more than $10.00?

```python
chipo_1 = chipo[chipo['item_price'] > '$10']
```

```python
chipo_1.count()
```

### Step 5. What is the price of each item? 
###### print a data frame with only two columns item_name and item_price

```python
chipo_1 = chipo_1[['item_name','item_price']]
```

```python
chipo_1.head()
```

### Step 6. Sort by the name of the item

```python
chipo_1.sort_values(by=['item_name'],axis = 0, ascending = [True], inplace = True )
```

```python
chipo_1
```

### Step 7. What was the quantity of the most expensive item ordered?

```python
chipo.sort_values(by=['quantity'],axis = 0, ascending = [False], inplace = True )
chipo
```

```python
chipo[chipo['item_name']=='Veggie Salad Bowl']['item_name'].count()
```
